### Data Visualization With Matplotlib
Matplotlib is a powerful data visualization library in Python that can be used to create a wide range of
plots and charts.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
plt.title("Basic line Plot")
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
plt.plot(x,y,color='green',linestyle='dotted',marker='o',linewidth=5)
plt.grid()
```

```python
x = [1,2,3,4,5]
y1 = [5,15,20,25,30]
y2 = [10,20,30,40,50]
plt.figure(figsize=(6,4))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("plot 1")
plt.figure(figsize=(6, 4))
plt.subplot(1, 2, 2)
plt.plot(x, y1, color='green')
plt.title("plot 2")
```

```python
# Bar plot

catagories = ['A','B','C','D','E']
values   = [1,3,4,7,9]
plt.bar(catagories,values)
```

### Sales Data Visulization

```python
import pandas as pd
import numpy as np
```

```python
sales_data = pd.read_csv('sales_data.csv')
print(sales_data.describe())
```

```python
total_sales_by_prduct = sales_data.groupby('Product Category')['Total Revenue'].sum()
```

```python
total_sales_by_prduct.plot(kind='bar',color='teal')
```

```python
sales_trend = sales_data.groupby('Date')['Total Revenue'].sum().reset_index()
```

```python
sales_trend
```

```python
sales_trend.plot()
```
